```python
import pickle
import matplotlib.pyplot as plt
from scipy.stats.mstats import gmean
import seaborn as sns
%matplotlib inline


price = pickle.load(open("price_record.p", "rb"))
```

```python
len(price[2])
```

```python
## Calculate price
```

```python

x = []
y =[]
for st, pr in price.items(): 
    #if step <=400: 
        x.append(st)
        y.append(gmean(pr))
        
y[0]
```

```python
fig, ax = plt.subplots(figsize = (7,7))
ax.scatter(x,y)
plt.title("Standard Approach: \nMean Trade Price", fontsize = 20, fontweight = "bold")
plt.xlabel("Time", fontsize = 15, fontweight = "bold")
plt.ylabel("Price", fontsize = 15, fontweight = "bold")
```

```python
x_vol = []
y_vol = []
total = 0
for s, p in price.items(): 
    #if step <=400:
        
        x_vol.append(s)
        y_vol.append(len(p))
        total += len(p)
total 
```

```python
fig2, ax2 = plt.subplots(figsize = (7,7))
ax2.hist(y_vol, 100)
plt.title("Trade Volume Histogram", fontsize = 20, fontweight = "bold")
plt.xlabel("Trade Volume of Step", fontsize = 15, fontweight = "bold")
plt.ylabel("Frequency Trade Volume", fontsize = 15, fontweight = "bold")
#plt.ylim(0,400)
```

```python
fig2, ax2 = plt.subplots(figsize = (7,7))
ax2.plot(x_vol, y_vol)
plt.title("Trade Volume", fontsize = 20, fontweight = "bold")
plt.xlabel("Time", fontsize = 15, fontweight = "bold")
plt.ylabel("Volume", fontsize = 15, fontweight = "bold")
ax2.text(600,300, "Total Trade Volume: \n "+str(total), fontsize = 15, fontweight = 'bold')
#plt.ylim(0,400)
```

```python
from statistics import stdev
from math import log
x_dev =[]
y_dev = []
x_all = []
y_all = []
log_prices = {}
for step, prices in price.items(): 
    log_prices[step] = [log(p) for p in prices]
for step, log_p in log_prices.items(): 
    #if step <= 400: 
        
        
        if len(log_p) <2: 
            pass       
        else: 
            for each in log_p: 
                x_all.append(step)
                y_all.append(each)
            
            x_dev.append(step)
            y_dev.append(stdev(log_p))
            
    
```

```python
from numpy.polynomial.polynomial import polyfit
fig3, ax3 = plt.subplots(figsize=(7,7))
ax3.scatter(x_all,y_all)
plt.plot(x_dev,y_dev,'-', color ='red')
plt.title("Standard Deviation of \nLogarithmic Mean", fontsize = 20, fontweight = "bold")
plt.xlabel("Time", fontsize = 15, fontweight = "bold")
plt.ylabel("Logarithmic Price", fontsize = 15, fontweight = "bold")
```
