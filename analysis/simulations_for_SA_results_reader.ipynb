```python
# initial houses

values = [0.001, 0.002, 0.003, 0.004, 0.005, 0.006, 0.007, 0.008, 0.009, 0.01]
outputs_price = []
outputs_contentment = []
error_price = []
error_contentment = []

counter = 0
for element in values:
    temp_outputs_price = []
    temp_outputs_contentment = []
    row = 1
    for i in range(0, 1000): # end range - 1
        row += 1
        if df_ofat2['num_houses'][i] == element:
            #counter += 1
            average_house_price = df_ofat2['average_house_price_Centrum'][i] + df_ofat2['average_house_price_Oost'][i] + df_ofat2['average_house_price_NieuwWest'][i] + df_ofat2['average_house_price_Zuidoost'][i] + df_ofat2['average_house_price_Noord'][i] + df_ofat2['average_house_price_West'][i] + df_ofat2['average_house_price_Zuid'][i]
            average_contentment = df_ofat2['average_contentment_Centrum'][i] + df_ofat2['average_contentment_Oost'][i] + df_ofat2['average_contentment_NieuwWest'][i] + df_ofat2['average_contentment_Zuidoost'][i] + df_ofat2['average_contentment_Noord'][i] + df_ofat2['average_contentment_West'][i] + df_ofat2['average_contentment_Zuid'][i]

            temp_outputs_price.append(average_house_price / 7)
            temp_outputs_contentment.append(average_contentment / 7)

    outputs_price.append(np.mean(temp_outputs_price))
    error_price.append(np.std(temp_outputs_price))
    outputs_contentment.append(np.mean(temp_outputs_contentment))
    error_contentment.append(np.std(temp_outputs_contentment))
    print(len(temp_outputs_contentment), 'samples')

plt.scatter(values, outputs_price)
plt.errorbar(values, outputs_price, yerr=error_price, fmt="o")
plt.xlabel('Initial houses')
plt.ylabel('Average house prices')
plt.show()

plt.scatter(values, outputs_contentment)
plt.errorbar(values, outputs_contentment, yerr=error_contentment, fmt="o")
plt.xlabel('Initial houses')
plt.ylabel('Average contentment')
plt.show()

# first 10 iterations are vatying initial houses; line 2 t/m 1001 [0.001, 0.002, 0.003, 0.004, 0.005, 0.006, 0.007, 0.008, 0.009 0.01]
# next 10 varyations are varying the noise; lines 1002 t/m 2001 []
# start_money_multiplier; 2002 t/m 3001
# start_money_multiplier_newcomers; 3002 t/m 4001
# contentment_threshold; lines 4002 t/m 5001
# weight_materialistic 5002 t/m 6001
# housing_growth_rate 6002 t/m 7001
# population_growth_rate  7002 t/m 8001

```
